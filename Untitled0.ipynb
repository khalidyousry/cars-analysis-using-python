```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cars.csv")
df.head(5)
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
# Group by Origin and calculate mean MPG(FUEL EFFICENCY)
regional_perf = df.groupby('Origin')['MPG'].mean().sort_values(ascending=False)
print(regional_perf)
```

```python
# Filter for 4, 6, and 8 cylinders and calculate means
engine_size_comp = df[df['Cylinders'].isin([4, 6, 8])].groupby('Cylinders')[['Horsepower', 'Displacement']].mean()
print(engine_size_comp)
```

```python
# 1. حساب سكور التسابق (Racing Score)
# كل ما زاد الـ Horsepower بالنسبة للوزن، كانت العربية "وحش" على الطريق
df['Racing_Score'] = (df['Horsepower'] / df['Weight']) * 100

# 2. حساب سكور العيلة (Family Score)
# هنا هنعتمد على كفاءة الوقود (MPG)
df['Family_Score'] = df['MPG']

# 3. دالة لتحديد التصنيف النهائي
def classify_car(row):
    if row['Racing_Score'] > df['Racing_Score'].quantile(0.8): # أعلى 20% في القوة
        return 'Racing Car'
    elif row['Family_Score'] > df['Family_Score'].median(): # أعلى من المتوسط في توفير البنزين
        return 'Family Car'
    else:
        return 'Standard/Classic'

df['Category'] = df.apply(classify_car, axis=1)

# عرض أفضل 5 عربيات لكل فئة
print("--- Top 5 Racing Cars ---")
print(df[df['Category'] == 'Racing Car'].sort_values(by='Racing_Score', ascending=False)[['Car', 'Racing_Score']].head())

print("\n--- Top 5 Family Cars ---")
print(df[df['Category'] == 'Family Car'].sort_values(by='Family_Score', ascending=False)[['Car', 'Family_Score']].head())
```

```python
# 2. Prepare Data for Questions
# Q4 Prep: Weight Category
df['Weight_Category'] = pd.qcut(df['Weight'], q=3, labels=['Light', 'Medium', 'Heavy'])

# Q5 Prep: Racing vs Family Classification
df['Racing_Score'] = (df['Horsepower'] / df['Weight']) * 100
df['Family_Score'] = df['MPG']
def classify(row):
    if row['Racing_Score'] > df['Racing_Score'].quantile(0.8):
        return 'Racing'
    elif row['Family_Score'] > df['Family_Score'].median():
        return 'Family'
    else:
        return 'Standard'
df['Category'] = df.apply(classify, axis=1)

# Function to safely remove legend to prevent the AttributeError
def safe_remove_legend(ax):
    leg = ax.get_legend()
    if leg:
        leg.remove()

# 3. Create Visualization Dashboard
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
plt.subplots_adjust(hspace=0.4)

# Plot 1: Regional Performance (Avg MPG)
regional_data = df.groupby('Origin')['MPG'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=regional_data, x='Origin', y='MPG', ax=axes[0, 0], palette='viridis', hue='Origin')
axes[0, 0].set_title('Avg MPG by Region (Fuel Efficiency)')
safe_remove_legend(axes[0, 0])

# Plot 2: Engine Size (Horsepower & Displacement)
engine_data = df[df['Cylinders'].isin([4, 6, 8])].groupby('Cylinders')[['Horsepower', 'Displacement']].mean().reset_index()
engine_melted = engine_data.melt(id_vars='Cylinders', var_name='Metric', value_name='Value')
sns.barplot(data=engine_melted, x='Cylinders', y='Value', hue='Metric', ax=axes[0, 1])
axes[0, 1].set_title('HP vs Displacement by Cylinder Count')

# Plot 3: Time Series (MPG Trend 1970-1982)
yearly_mpg = df.groupby('Model')['MPG'].mean().reset_index()
sns.lineplot(data=yearly_mpg, x='Model', y='MPG', marker='o', ax=axes[1, 0], color='red')
axes[1, 0].set_title('Average MPG Trend (1970 - 1982)')
axes[1, 0].grid(True)

# Plot 4: Weight vs Acceleration
weight_accel = df.groupby('Weight_Category', observed=False)['Acceleration'].mean().reset_index()
sns.barplot(data=weight_accel, x='Weight_Category', y='Acceleration', ax=axes[1, 1], palette='magma', hue='Weight_Category')
axes[1, 1].set_title('Avg Acceleration (Lower is Faster) by Weight Class')
safe_remove_legend(axes[1, 1])

# Plot 5: Racing vs Family Classification
sns.scatterplot(data=df, x='MPG', y='Horsepower', hue='Category', style='Category', ax=axes[2, 0], palette='Set1', s=100)
axes[2, 0].set_title('Racing vs Family Car Classification')

# Plot 6: Correlation Matrix (Heatmap)
corr = df[['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[2, 1])
axes[2, 1].set_title('Correlation Matrix of Vehicle Stats')

plt.tight_layout()
plt.show()
```
